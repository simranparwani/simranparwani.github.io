```python
import pandas as pd
import numpy as np
pd.options.display.max_rows = 500
```

```python
oh2004 = pd.read_csv('oh2004.tsv', sep='\t')
oh2014 = pd.read_csv('oh2014.tsv', sep='\t')
oh2016 = pd.read_csv('oh2016.tsv', sep='\t')
oh2019 = pd.read_csv('oh2019.tsv', sep='\t')
```

```python
count2004=oh2004.groupby(oh2004['county']).count()['newspaper_name']
```

```python
count2014=oh2014.groupby(oh2014['county']).count()['newspaper_name']
count2016=oh2016.groupby(oh2016['county']).count()['newspaper_name']
count2019=oh2019.groupby(oh2019['county']).count()['newspaper_name']
```

```python
create_frame = {'2004': count2004, '2014': count2014, '2016': count2016, '2019':count2019}
num_newspapers = pd.DataFrame(create_frame)
```

```python
num_newspapers.fillna(0, inplace=True)
num_newspapers = num_newspapers.reset_index()
```

```python
num_newspapers.columns =  ['county', '2004', '2014', '2016', '2019']
```

```python
num_newspapers['county']=num_newspapers['county'].apply(lambda x: x + " County")
```

```python
num_newspapers.to_csv('newsdeserts.csv', index=False)
```

```python
year2004 = oh2004.drop(['frequency', 'newspaper_name', 'county', 'state','city', 'owner_id', 'owner_name', 'owner_type', 'days_published'], axis=1)
year2004 = year2004.rename(columns={'total_circulation': '2004'})

year2014 = oh2014.drop(['frequency', 'newspaper_name', 'county', 'state','city', 'owner_id', 'owner_name', 'owner_type', 'days_published'], axis=1)
year2014 = year2014.rename(columns={'total_circulation': '2014'})

year2016 = oh2016.drop(['frequency', 'newspaper_name', 'county', 'state','city', 'owner_id', 'owner_name', 'owner_type', 'days_published'], axis=1)
year2016 = year2016.rename(columns={'total_circulation': '2016'})

year2019 = oh2019.drop(['frequency', 'newspaper_name', 'county', 'state','city', 'owner_id', 'owner_name', 'owner_type', 'days_published'], axis=1)
year2019 = year2019.rename(columns={'total_circulation':  '2019'})
```

```python
circulation = year2004.set_index('newspaper_id').join([year2014.set_index('newspaper_id'),year2016.set_index('newspaper_id'), year2019.set_index('newspaper_id')], how='outer')
```

```python
circulation
```

```python
circulation = circulation.reset_index()
circulation['newspaper_id'].value_counts().head(5)
```

```python
circulation['2004']=circulation['2004'].str.replace(',', '').str.replace('-','')
circulation['2014']=circulation['2014'].str.replace(',', '').astype(float)
circulation['2016']=circulation['2016'].str.replace(',', '').astype(float)
```

```python
circulation['2004'] = circulation['2004'].astype(float)
```
