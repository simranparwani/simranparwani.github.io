```python
import pandas as pd
pd.options.display.max_rows = 100
```

```python
oh2004 = pd.read_csv('oh2004.tsv', sep='\t')
oh2014 = pd.read_csv('oh2014.tsv', sep='\t')
oh2016 = pd.read_csv('oh2016.tsv', sep='\t')
oh2019 = pd.read_csv('oh2019.tsv', sep='\t')
```

```python
count2004=oh2004.groupby(oh2004['county']).count()['newspaper_name']
```

```python
count2014=oh2014.groupby(oh2014['county']).count()['newspaper_name']
count2016=oh2016.groupby(oh2016['county']).count()['newspaper_name']
count2019=oh2019.groupby(oh2019['county']).count()['newspaper_name']
```

```python
create_frame = {'2004': count2004, '2014': count2014, '2016': count2016, '2019':count2019}
num_newspapers = pd.DataFrame(create_frame)
```

```python
num_newspapers.fillna(0, inplace=True)
num_newspapers = num_newspapers.reset_index()
```

```python
num_newspapers.columns =  ['county', '2004', '2014', '2016', '2019']
```

```python
num_newspapers['county']=num_newspapers['county'].apply(lambda x: x + " County")
```

```python
num_newspapers.to_csv('newsdeserts.csv', index=False)
```
